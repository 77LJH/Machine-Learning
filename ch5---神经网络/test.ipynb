```python
import matplotlib.pyplot as plt
import numpy as np

# 
plt.figure(figsize=(10, 4)) # Figure104


# 
plt.subplot(1, 2, 1)  #  1 1  2 2   1 1
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
plt.plot(x1, y1, label='Sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Function')

# 
plt.subplot(1, 2, 2)  # 1  2 
x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)
plt.plot(x2, y2, color='red', label='Cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cosine Function')

# 
plt.legend()

# 
plt.show()
```
