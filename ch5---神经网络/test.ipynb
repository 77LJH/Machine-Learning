```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个包含多个子图的图表
plt.figure(figsize=(10, 4)) # 创建一个新的图表对象（Figure），并设置图表的大小为宽度为10单位、高度为4单位


# 第一个子图
plt.subplot(1, 2, 1)  # 第一个参数 1 表示将图表分割成1行 第二个参数 2 表示将图表分割成2列  第三个参数 1 表示当前子图位于第1个位置，也就是图表的左边
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
plt.plot(x1, y1, label='Sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Function')

# 第二个子图
plt.subplot(1, 2, 2)  # 1 行 2 列，第二个子图
x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)
plt.plot(x2, y2, color='red', label='Cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cosine Function')

# 显示图例
plt.legend()

# 显示图表
plt.show()
```
